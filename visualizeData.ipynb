```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('100g4dfitnessOverTime.csv')

plt.rcParams['text.usetex'] = False
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

# plot data
plt.figure(figsize=(10, 6))
plt.ylim(bottom=-4, top=4)
plt.xlim(0, 100)
plt.plot(df.index, df['averageFitness'], marker='.', color='black')
plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.grid(True, linestyle='--', alpha=0.7)

# line of best fit
coeff = np.polyfit(df.index, df['averageFitness'], 1)
lm = np.poly1d(coeff)
slope = coeff[0]
b = coeff[1]
equation = f'y = {slope:.2}x + {b:.2}'

plt.plot(df.index, lm(df.index), color='red', linestyle='-', label='Line of Best Fit')
plt.text(0.05, 0.95, f'{equation}',
         transform=plt.gca().transAxes,
         verticalalignment='top')

plt.show()
```
